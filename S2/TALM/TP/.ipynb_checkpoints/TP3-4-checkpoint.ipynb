```python
import re 

FILE_PATH = "ai_text.txt" 

with open(FILE_PATH, "r", encoding="utf-8") as f: 
    AI_TEXT = f.read()

docs = [p.strip() for p in AI_TEXT.strip().split("\n\n") if p.strip()] 
```

```python
def whitespace_tokenise(text: str):
    return text.split()

def rule_based_tokenise(text: str):
    return re.findall(r"\w+|[^\w\s]", text)
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))
negations_to_keep = {"no", "not", "nor", "never"} 
stop_words = stop_words - negations_to_keep 
def remove_stopwords(tokens, stopset):
         return [t for t in tokens if t.lower() not in stopset]
```

```python
def custom_analyser(text: str):
    tokens = rule_based_tokenise(text)
    tokens = [t.lower() for t in tokens if t.isalpha()]
    tokens = remove_stopwords(tokens, stop_words)
    return tokens

def custom_analyser_with_bigrams(text: str):
    tokens = custom_analyser(text)
    bigrams = [f"{tokens[i]} {tokens[i+1]}" for i in range(len(tokens)-1)]
    return tokens + bigrams
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv_uni = CountVectorizer(analyzer=custom_analyser, ngram_range=(1,1))
X_uni = cv_uni.fit_transform(docs)
df_uni = pd.DataFrame(X_uni.toarray(), columns=cv_uni.get_feature_names_out())

cv_12 = CountVectorizer(analyzer=custom_analyser_with_bigrams)
X_12 = cv_12.fit_transform(docs)
df_12 = pd.DataFrame(X_12.toarray(), columns=cv_12.get_feature_names_out())
```

```python
from sklearn.preprocessing import normalize
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

X_uni_l2 = normalize(X_uni, norm="l2")
X_12_l2 = normalize(X_12, norm="l2")

sim_uni = cosine_similarity(X_uni_l2)
sim_12 = cosine_similarity(X_12_l2)
```

```python
def custom_analyser(text: str):
    """Analyzer for unigrams only"""
    tokens = rule_based_tokenise(text)
    tokens = [t.lower() for t in tokens if t.isalpha()]
    tokens = remove_stopwords(tokens, stop_words)
    tokens = lemmatise_tokens(tokens)
    return tokens

def custom_analyser_with_bigrams(text: str):
    """Analyzer that generates both unigrams and bigrams"""
    tokens = custom_analyser(text)
    bigrams = [f"{tokens[i]} {tokens[i+1]}" for i in range(len(tokens)-1)]
    return tokens + bigrams

print("Unigrams (first 30):")
print(custom_analyser(docs[0])[:30])
print("\nWith bigrams (first 40):")
print(custom_analyser_with_bigrams(docs[0])[:40])
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score

docs = [
    "AI improves machine learning systems",
    "Neural networks are powerful models",
    "Support vector machines are supervised",
    "AI ethics is important for society",
    "Accountability in AI systems matters",
    "Ethical AI requires transparency"
]

y = np.array([0, 0, 0, 1, 1, 1])

cv_12 = CountVectorizer(ngram_range=(1,2))
X_12 = cv_12.fit_transform(docs)

X_train, X_test, y_train, y_test = train_test_split(
    X_12, y, test_size=0.33, random_state=42, stratify=y
)

clf_nb = MultinomialNB()
clf_nb.fit(X_train, y_train)
y_pred_nb = clf_nb.predict(X_test)

clf_svm = SVC(kernel='linear')
clf_svm.fit(X_train, y_train)
y_pred_svm = clf_svm.predict(X_test)

print("Naive Bayes Report:")
print(classification_report(y_test, y_pred_nb, zero_division=0))
print("Accuracy:", accuracy_score(y_test, y_pred_nb))

print("\nSVM Report:")
print(classification_report(y_test, y_pred_svm, zero_division=0))
print("Accuracy:", accuracy_score(y_test, y_pred_svm))

new_text = ["AI ethics and accountability are important for the future."]
new_text_transformed = cv_12.transform(new_text)
prediction = clf_nb.predict(new_text_transformed)
label = "Ethique" if prediction[0] == 1 else "Technique"

print("\nTexte:", new_text[0])
print("Prediction:", label)
```
