```python
with open("sample.txt", "r") as file: text = file.read()
print(text)
```

```python
tokens = text.split()
print(tokens)
```

```python
import re
simple_text = text
simple_pattern = r"imad"
result = re.search(simple_pattern, simple_text)
print(result.group())
```

```python
email_pattern = r"(\S+@\S+\.\S+)"
email_result = re.search(email_pattern, text)
if email_result: print("Email:", email_result.group(1))
else: print("No email found")
```

```python
tel_pattern = r"\b\d{10}\b"
tel_result = re.search(tel_pattern, text)
if tel_result:print("Telephone:", tel_result.group())
else:print("No telephone found")
```

```python
date_pattern = r"(\d{2}/\d{2}/\d{4})"
date_result = re.search(date_pattern, text)
if date_result: print("Birth Date:", date_result.group())
else: print("No birth date found")
```
