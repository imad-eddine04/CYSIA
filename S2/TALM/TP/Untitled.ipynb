```python
# File path (same folder as your notebook is easiest)
FILE_PATH = "ai_text.txt"

with open(FILE_PATH, "r", encoding="utf-8") as f:
    AI_TEXT = f.read ()

print ("Number of characters:", len(AI_TEXT))
print("Preview:\n", AI_TEXT[:400]," ... ")
```

```python
import re
docs = [p.strip() for p in AI_TEXT.strip().split("\n\n") if p.strip()]
print("Number of documents (paragraphs):", len(docs))
for i, d in enumerate(docs, 1):
    print(f"\n--- Document {i} preview ---\n{d[:250]}...")
```

```python
def whitespace_tokenise(text: str):
    return text.split()
print("Whitespace tokens (first 40):")
print(whitespace_tokenise(docs [0]) [:40])
```

```python
def rule_based_tokenise(text: str):
    return re.findall(r"\w+|[^\w\s]", text)
print("Rule-based tokens (first 60):")
print (rule_based_tokenise (docs [0])[:60])
```

```python
import nltk
from nltk. corpus import stopwords

# One-time download (run once if needed)
# nltk. download ("stopwords")

stop_words = set(stopwords.words("english"))

# Recommended: keep negations
negations_to_keep = {"no", "not", "nor", "never"}
stop_words = stop_words - negations_to_keep

def remove_stopwords(tokens, stopset) :
    return [t for t in tokens if t. lower() not in stopset]

tokens_demo = [t.lower() for t in rule_based_tokenise(docs [0]) if t.isalpha()]
tokens_no_sw = remove_stopwords (tokens_demo, stop_words)

print ("Before stopwords:", tokens_demo [:30])
print ("After stopwords :", tokens_no_sw[:30])
```
