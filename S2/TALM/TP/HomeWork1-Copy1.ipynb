```python
with open("Algerian.txt", "r") as file:
    text = file.read()
print(text)
```

```python
import re
#extract ooredoo numbers
mobilis_pattern = r"(\+213\s6\s\d{2}\s\d{2}\s\d{2}\s\d{2})"
mobilis_result = re.findall(mobilis_pattern, text)
if mobilis_result:
    for i in mobilis_result:
        print("mobilis phone : ", i)
else: print("Mobilise phones Not Found!")
#extract ooredoo numbers
ooredoo_pattern = r"(\+213\s5\s\d{2}\s\d{2}\s\d{2}\s\d{2})"
ooredoo_result = re.findall(ooredoo_pattern, text)
if ooredoo_result: 
    print(f"Ther is {len(ooredoo_result)} ooredoo Phones : ")
    for i in ooredoo_result:
        print("ooredoo phone : ", i)
else: print("ooredoo phones Not Found!")
```

```python
#now for the emails from aintmouchen
tmouch_pattern = r"(\S+@univ-temouchent.edu.dz)"
tmouch_result = re.findall(tmouch_pattern, text)
if tmouch_result: 
    for i in tmouch_result:
        print("tmouchent mails : ", i)
else: print("tmouchent mails Not Found!")
```
